```python
import os
import netCDF4

import scipy
import numpy as np
import pandas as pd
import seaborn as sns

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature

from matplotlib.collections import LineCollection
from matplotlib.path import Path
import matplotlib.path as mpath
import matplotlib.patches as patches
```

```python
#################
# Daily netcdfs Z 1
#################
#Path
os.chdir('D:\\Bigfiles\\ERA5_Feb2020_Attribution')
filename_pres_1 = 'TFeb2020_Attr_pres-1950-1978_3h_Z500.nc'

#Load netcdf
ncdf_1 = netCDF4.Dataset(filename_pres_1)

#Get dimensions
lons = ncdf_1.variables['longitude'][:]
lats = ncdf_1.variables['latitude'][:]
times_1 = ncdf_1.variables['time'][:]


#Mean daily values 1950-1978
z = ncdf_1.variables['z'][:,:,:]
times_24H = np.empty(int(len(times_1)/8))
z_24H = np.empty((int(len(times_1)/8), len(lats), len(lons)))
for i in range(np.size(times_1[::8])):
    print(i)
    times_24H[i] = times_1[i*8]
    z_24H[i,:,:] = np.mean(z[i*8:i*8+8,:,:], axis=0)  
    
#Write numpy array as NetCDF
ncout = netCDF4.Dataset('Feb2020_Attr_pres-1950-1978_Z500_24H.nc','w','NETCDF4_classic'); # using netCDF3 for output format 
ncout.createDimension('time',np.shape(z_24H)[0])
ncout.createDimension('longitude',np.shape(z_24H)[2])
ncout.createDimension('latitude',np.shape(z_24H)[1])

timevar = ncout.createVariable('time','float64',('time')); timevar.setncattr('units','hours since 1900-01-01 00:00:00.0'); timevar[:]=times_24H;

longitude = ncout.createVariable('longitude','float32',('longitude')); longitude[:] = ncdf_1.variables['longitude'][:];
latitude = ncout.createVariable('latitude','float32',('latitude')); latitude[:] = ncdf_1.variables['latitude'][:];

z_join = ncout.createVariable('z','float32',('time','latitude','longitude')); z_join.setncattr('units','m'); z_join[:] = z_24H;

ncout.close();
```
